```python
import nltk
from nltk.text import Text
from nltk.tokenize import word_tokenize
with open("corpus/threebears.txt", "r", encoding="utf-8") as file:
    raw_text = file.read()
tokens = word_tokenize(raw_text)
text_obj = Text(tokens)
text_obj.concordance('Wee')
```
